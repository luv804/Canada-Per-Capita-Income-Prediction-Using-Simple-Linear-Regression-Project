```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
#Loading our dataset
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

### Data Cleaning

```python
df.isnull()
```

```python
df.isnull().sum()
```

### Data Analysis

```python
df.describe()
```

```python
df.info()
```

```python
# plotting scatter plot to get idea of our data points
plt.xlabel("year")
plt.ylabel("per capita income(US$)")
plt.scatter(df.year,df['per capita income (US$)'].values,color="red",marker="+")
```

### Model Training

```python
# Data Selection
X = df[['year']].values
y = df['per capita income (US$)'].values
X
```

```python
y
```

```python
# Splitting the dataset into training data (80%) and testing data (20%)
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
# creating our Linear Regression object
model = LinearRegression()
```

```python
# fitting the training data into our LinearRegression model
model.fit(X_train,y_train)
```

```python
model.coef_ # The value of m for fitting the best line (y = mx + b)
```

```python
model.intercept_ # The value of b for fitting the best line (y = mx + b)
```

```python
# Values of y predicted by the model from training data
y_predicted_train = model.predict(X_train)
```

```python
# Plotting our LinearRegression Line
plt.xlabel("year")
plt.ylabel("per capita income(US$)")
plt.scatter(X_train,y_train,color="red",marker="+")
plt.plot(X_train,y_predicted_train,color="darkblue")
```

### Evaluating Model Performance on Test Data

```python
model.score(X_test,y_test)
```

```python
# Predicted Values of y by our model on Testing Data
y_predicted_test = model.predict(X_test)
```

```python
# Plotting our LinearRegression Line for Testing Data
plt.xlabel("year")
plt.ylabel("per capita income(US$)")
plt.scatter(X_test,y_test,color="red",marker="+")
plt.plot(X_test,y_predicted_test,color="darkblue")
```

### Predicting Per Capita Income

```python
pred = model.predict([[2040]])
print("Canada Per Capita Income Prediction : ")
print("Year : 2040")
print("Predicted Per capita income (US$) : ",round(pred[0],2))
```
